```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import ActionChains
import time
#import csv

#from selenium import webdriver
from selenium.webdriver.chrome.service import Service

service = Service(executable_path = "D:/chromedriver4.exe")
driver = webdriver.Chrome(service=service)

list0 = [] 
list1 = []
list2 = []
list3 = []
list4 = [] 

i = 0 

#path = 'D:/chromedriver.exe'
#driver = webdriver.Chrome(path) 
#查詢頁面網址
driver.get('https://c.8891.com.tw/Models') 

#選SUV
driver.find_element(By.XPATH, '//*[@id="findCar"]/div/div[2]/div[2]/div[1]/div[2]/div[2]/div/div[2]').click()  
time.sleep(1)

#選價位120~180
driver.find_element(By.XPATH, '//*[@id="findCar"]/div/div[2]/div[2]/div[1]/div[2]/div[1]/div/div[5]').click()  
time.sleep(1)

#-------------------------------------------------
# 等{在售}出來
try :  
  WebDriverWait(driver, 10).until(
      EC.presence_of_element_located((By.XPATH, '//*[@id="findCar"]/div/div[2]/div[4]/div[1]/div[1]/ul/li[1]'))
  )
except : 
   print("kiki")

time.sleep(3)

#選價格低優先
driver.find_element(By.XPATH, '//*[@id="findCar"]/div/div[2]/div[4]/div[1]/div[1]/div/label[3]').click()    

time.sleep(2)

#向下滑動
js = "window.scrollTo(0, document.body.scrollHeight);"
driver.execute_script(js) 


#下一頁
driver.find_element(By.XPATH, '//*[@id="findCar"]/div/div[2]/div[4]/div[1]/div[3]/ul/li[4]').click() 
time.sleep(3)


#蒐集連結
car_name = driver.find_elements(By.CLASS_NAME, 'item-msg--uHn6z')

for item in car_name :
   url = item.get_attribute("href")
   list0.append(url)
   #print('連結:', list0)

#----------------------------------------------

k = 0 
for i in range(len(list0)) :
   
   driver.get(list0[k]) 
   k = k + 1 

   #車名
   titles = driver.find_elements(By.CLASS_NAME, "fl.summary-top-name")

   #車型
   types = driver.find_elements(By.CLASS_NAME, "content--YQ5Go")

   #形式
   power = driver.find_elements(By.CLASS_NAME, "info--AZQyE")

   #oil
   oil = driver.find_elements(By.CLASS_NAME, "info--AZQyE")
   
   
   for title in titles :  
      name = title.text
      list1.append(name)

   for title in types :  
      name1 = title.text
      list2.append(name1)

   for title in power :  
      name2 = title.text
      list3.append(name2)

   for title in oil :  
      name3 = title.text
      list4.append(name3)


print(list1)  
print(list2) 
print(list3)
#print(list4) 

time.sleep(3)

driver.quit()
```
